#### Comparaciones lógicas


```python
print ("Comparaciones lógicas")
a = 4
b = 7

# a mayor que b
print (f"{a}>{b}",a>b)

# b mayor o igual que a
print (f"{b}<={a}",b<=a)


# b igual que 4
print (f"{b}==2",b==2)


# a es diferente 5
print (f"{a}!=5",a!=5)

# obtener el opuesto de un resultado no logico

print(f"not {a}!=5",not a !=5)

#Evaluar una condicion con and

print(f"{a}!=5 and {a}>{b}",a!= 5 and a>b)

# Evaluar si se cumple una condición o otra

print(f"{a}!=5 or {b}<={a}",a!= 5 or b<= a)
```

```python
while 1:
    numero = input ("indica un numero del 1 al 5, o escribe 'done' para salir:")
    try:
        if numero == "done":
            print("Good Look")
            break
        elif int(numero)<= 5: #controla que que los numeros sean del 1 al 5
            cubo = int(numero)**3
            print("El número {} al cubo es: {}".format(numero,cubo))
        else:
            print("El dato {} ingresado no es un número".format(numero))
    except ValueError:
        print("Dato ingresado no valido!")
```

### Funciones para devolver un valor o operación:

```python
def cuadrados(x, y):
    return x**2 + y**2 # la función finaliza con el return, si se escribe otras lineas despúes no se ejecutan

result = cuadrados(7,4)

print(result)
```

Con el resultado se puede invocar nuevamente la función y enviandola como argumento

```python
result = cuadrados(result,35)

print(result)
type(result)
```

Las funciones pueden devolver varios resultados y las devuelve como **Tuplas**

En este ejemplo se envia dos argumentos y se invoca la funcion sal_comi, ella tiene dos variables
que devuelven la multiplicacion de x,y y la multiplicacion del resultado de la variable sal
por una costante de 0.30

```python
def sal_comi (x,y):
    sal = x * y
    com = sal * 0.30
    return sal, com

total = sal_comi(10000,80)

print(total)
print(type(total))
```

El resultado de esta tupla se puede dividir y asi manejar dos tuplas diferentes

```python
salario, comision = sal_comi(10000,80)
print(salario)
print(comision)
```

## COUNTER

Es una clase que permite contar elementos, se define de dos formas:
    

```python
# Forma basica

import collections

collections.Counter
```

```python
# Forma avanzada

from collections import counter


ordenes = ["f-1516",
           "f-1516",
           "f-1517",
           "f-1516",
           "f-1516",
           "f-1518",
           "f-1516"
          ]
cantidad_or= Counter(ordenes)

print(cantidad_or)
```

```python
#Añadir elementos a un Counter de forma dinámica

print(cantidad_or["f-1518"])

cantidad_or.update(["f-1518"])

print(cantidad_or["f-1518"])
```

```python
# Aumentar el conteo mediante un diccionario

cantidad_or.update({"f-1517":5,"f-1518":6})

print (cantidad_or)
```
